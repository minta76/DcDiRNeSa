```python
import random
def getPair(Rlist):
    random_num1 = random.choice(Rlist)
    random_num2 = random.choice(Rlist)
    while random_num1==random_num2:
        random_num2 = random.choice(Rlist)
    return sorted([random_num1,random_num2])
```

```python
#create negatives samples
negSam=[]
sampleSize=len(inp)
while len(negSam)<sampleSize:
    pair=getPair(list(inter))
    while pair in Two_comb.values() and pair[0] not in list(df.index) and pair[1] not in list(df.index):
       pair=getPair(list(inter))
    negSam.append(pair)
print(len(negSam))
```

```python
#create negetive input on NN
neginp=[]
for comb in negSam:
    #print(comb[0],comb[1])
    if comb[0] in list(df.index) and comb[1] in list(df.index):
        neginp.append([n for n in ast.literal_eval(df.loc[comb[0],'vector'])]+[n for n in ast.literal_eval(df.loc[comb[1],'vector'])])
print(len(neginp))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import preprocessing

X = inp+neginp
y =  [1] * len(inp)+[0]*len(neginp)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)
mlp = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)


scaler = preprocessing.StandardScaler()#.fit(X_train)
vectors = scaler.fit_transform(X_train)
vectors_test = scaler.transform(X_test)
mlp.fit(vectors, y_train)
predict_test = mlp.predict(vectors_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
scalar = StandardScaler()
from sklearn.model_selection import KFold
from sklearn.model_selection import ShuffleSplit
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score

print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))

scalar = StandardScaler()
pipeline = Pipeline([('transformer', scalar), ('estimator', mlp)])

cv =ShuffleSplit(n_splits=10, test_size=0.3, random_state=0)# KFold(n_splits=4)
scores = cross_val_score(pipeline, X, y, cv = cv,scoring='f1')
print(scores)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```
